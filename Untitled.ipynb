```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

os.listdir()
```

```python
weather = pd.read_csv('timberline_cleaned_weather.csv', index_col='Date/Time (PST)')
forecasts = pd.read_csv('forecasts_no_dups.csv', index_col='Unnamed: 0')
```

```python
forecasts.head(2)
```

```python
combined = pd.concat([weather, forecasts], axis=1)
```

```python
combined.drop(combined.index[:133], inplace=True)
```

```python
combined.Snow_Depth_tl.clip(0, 200, inplace=True)
combined.Snow_Depth_tl.fillna(method='backfill', inplace=True)
combined.Snow_Depth_tl.plot()
```

```python
combined.Snow_Depth_tl = combined.Snow_Depth_tl.rolling(center=True, window=10).median()
```

```python
combined.Snow_Depth_tl.plot()
```
